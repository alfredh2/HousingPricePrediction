# Using Machine Learning to predict housing prices in California

### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load Data

```python
df = pd.read_csv('housing.csv')
```

```python
df
```

```python
# Check for any null values in the dataset
df.info()
```

```python
# This drops the rows containing any null values
df = df.dropna()
```

```python
df.info()
```

### Split the data

```python
from sklearn.model_selection import train_test_split

# X is everything but the target column, y is just the target
X = df.drop(['median_house_value'], axis=1)
y = df['median_house_value']
```

```python
# Split with a 20% test set size
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
train_data = X_train.join(y_train)
```

```python
X_test
```

```python
# Train data visualization
train_data.hist(figsize=(15,8))
plt.show()
```

```python
# Heat map to see which features are most correlated
train_data_num = train_data.drop('ocean_proximity', axis=1)
plt.figure(figsize=(15,8))
sns.heatmap(train_data_num.corr(), annot=True, cmap="YlGnBu")
plt.show()
```

### Preprocessing

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)

train_data.hist(figsize=(16,8))
plt.show()
```

```python
# One hot encoding for ocean proximity
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop('ocean_proximity', axis=1)
train_data
```

```python
# Heat map to see which features are most correlated
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude', y='longitude', data=train_data, hue='median_house_value', palette='coolwarm')
plt.show()
```

### Feature Engineering

```python
# Create new features for bedroom ratio and household rooms
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
plt.show()
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)
model = LinearRegression()
model.fit(X_train_s, y_train)
```

```python
# Next time split after all this
test_data = X_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop('ocean_proximity', axis=1)

test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']

# Add missing columns in test set (Not always an island instance when one hot encoding test set)
for column in train_data.columns:
    if column not in test_data.columns:
        test_data[column] = 0

# Reorder test columns according to train columns
test_data = test_data[train_data.columns]
```

```python
X_test, y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
X_test_s = scaler.transform(X_test)
```

```python
model.score(X_test_s, y_test)
```

### Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor()
forest.fit(X_train_s, y_train)
forest.score(X_test_s, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Random forest model with hyperparameter tuning
forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [100, 200, 300],
    "min_samples_split": [2, 4],
    "max_depth": [None, 4, 8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring="neg_mean_squared_error", return_train_score=True)

grid_search.fit(X_train_s, y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(X_test_s, y_test)
```
